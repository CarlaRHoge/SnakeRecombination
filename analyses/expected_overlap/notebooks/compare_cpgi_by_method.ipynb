```python
d = pd.DataFrame()
islands = {"corn":["cpgplot", "hmm"],"rattle":["cpgplot_default", "cpgplot_schield"]}

for sp in ["corn", "rattle"]:
    for isl in islands[sp]:
        df = pd.read_csv("../cpgi_copy/{}.{}.bed.gz".format(sp, isl), 
                         sep="\t", header=None)
        df.columns = ["scaf", "start", "end"]
        df["sp"] = sp
        df["isl"] = isl
        df["width"] = df.end - df.start
        d = pd.concat([d, df])
```

```python
g = sns.displot(data=d, row="sp", x="width", hue="isl", kind="ecdf")
g.set(xscale="log")
```
