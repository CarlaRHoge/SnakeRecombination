```python
d = pd.read_csv("../keep_shuf/annotated_shufs.bed.gz",sep="\t", header=None)
```

```python
with open("../misc_files/repeats_aboveN500.txt", "r") as fh:
    repeats = [line.strip().replace("/",".").replace(".","_").replace("-","_") for line in fh]
```

```python
repeats_c = ["{}_c".format(p) for p in repeats]
```

```python
d.columns = ["scaf", "start", "end", "cross_id", 
             "cpgi", "prdm9", "hotspot","rep", 
             "fam", "allele"] + sum([["{}_c".format(p),p] for p in repeats], [])
```

```python
d["cbool"] = d.rep==0
d["width"] = d.end - d.start
```

```python
subd = d[(d.cpgi>5e3) & (d.width<4e3)]
```

```python
fold_df = subd.groupby("cbool")[repeats].mean().transpose().reset_index()
fold_df.columns = ["repeat", "noncrossover", "crossover"]
fold_df["fold"] = fold_df.noncrossover/fold_df.crossover
```
